```python
#Here is the request library that is used to get the 
import requests
#Here is the link that is opeing for the scraping
URL = "https://www.bnwcollections.com/products/canon-eos-77d-dslr-camera-body-only-price-in-pakistan"
#Here i am getting the link by using the request library and store the HTML into a variable
page = requests.get(URL)
#Here i am converting the HTML into a text format
print(page.text)
```

```python
#Here is the request library that is used to get the 
import requests
#here is the beautifulsoup library that is used to get the
from bs4 import BeautifulSoup
#Here is the link that is opeing for the scraping
URL = "https://www.bnwcollections.com/products/canon-eos-77d-dslr-camera-body-only-price-in-pakistan"
#Here i am getting the link by using the request library and store the HTML into a variable
page = requests.get(URL)
#here is yhe beautiful soup library and get the page content
soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup)
```

```python
#here is the soup get the title
print(soup.title)
```

```python
#here is the soup get the title name
print(soup.title .name)
```

```python
#get the title name
print(soup.title.string)
```

```python
#to get the parent name
print(soup.title.parent.name)
```

```python
#to print paragraph
print(soup.p)
```

```python
#anchor tag a
print(soup.a)
```

```python
#find all complete anchor a
print(soup.find_all('a'))
```

```python
# one anchor tag select
print(soup.find('a'))
```

```python
#find all anchor tag to  get the link 
for link in soup.find_all('a'):
  print(link.get('href'))
```
